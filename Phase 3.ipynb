## Feature Engineering ##

```python
# Define function to get BMI category
def get_bmi_category(bmi):
    if bmi <= 18.5:
        return 'Underweight'
    elif bmi <= 24.9:
        return 'Normal weight'
    elif bmi <= 29.9:
        return 'Overweight'
    else:
        return 'Obesity'

# Apply the function to create BMI_Category column
df_new['BMI_Category'] = df_new['BMI'].apply(get_bmi_category)

# Health Checkup Frequency
checkup_mapping = {'Within the past year': 4, 'Within the past 2 years': 2, 'Within the past 5 years': 1, '5 or more years ago': 0.2, 'Never': 0}
df_new['Checkup_Frequency'] = df_new['Checkup'].replace(checkup_mapping)

# Lifestyle Score
exercise_mapping = {'Yes': 1, 'No': 0}
smoking_mapping = {'Yes': 1, 'No': 0}
df_new['Lifestyle_Score'] = df_new['Exercise'].replace(exercise_mapping) - df_new['Smoking_History'].replace(smoking_mapping) + df_new['Fruit_Consumption']/10 + df_new['Green_Vegetables_Consumption']/10 - df_new['Alcohol_Consumption']/10

# Healthy Diet Score
df_new['Healthy_Diet_Score'] = df_new['Fruit_Consumption']/10 + df_new['Green_Vegetables_Consumption']/10 - df_new['FriedPotato_Consumption']/10

# Interaction Terms
df_new['Smoking_Alcohol'] = df_new['Smoking_History'].replace(smoking_mapping) * df_new['Alcohol_Consumption']
df_new['Checkup_Exercise'] = df_new['Checkup_Frequency'] * df_new['Exercise'].replace(exercise_mapping)

# Ratio of Height to Weight
df_new['Height_to_Weight'] = df_new['Height_(cm)'] / df_new['Weight_(kg)']

# Healthy_Diet_Lifestyle Interaction

df_new['HealthyDiet_Lifestyle'] = df_new['Healthy_Diet_Score'] * df_new['Lifestyle_Score']
df_new.head()
```

### Why we did Featured Engineering
BMI Category: We calculated and assigned a categorical label to each individual based on their BMI. This categorization can be more informative than using the raw BMI value alone, as it provides a clear understanding of whether the individual falls into the underweight, normal weight, overweight, or obesity category.

Health Checkup Frequency: We map the values of the 'Checkup' column to numerical values that represent the frequency of health checkups. This allows the model to capture the impact of how frequently an individual undergoes health checkups on their risk of cardiovascular disease.

Lifestyle Score: We calculate a composite score that takes into account various lifestyle factors such as exercise, smoking history, fruit consumption, green vegetables consumption, and alcohol consumption. This score provides a more holistic representation of an individual's lifestyle habits, which can help the model capture the combined effect of these factors on cardiovascular disease risk.

Healthy Diet Score: Similar to the lifestyle score, the healthy diet score is calculated based on an individual's fruit consumption, green vegetables consumption, and fried potato consumption. This score reflects the quality of an individual's diet in terms of healthy food choices.

Interaction Terms: Interaction terms are created by multiplying different variables together. For instance, the 'Smoking_Alcohol' term is the result of multiplying smoking history and alcohol consumption. By combining different variables in the model, we're checking if they work together to either boost each other's impact or make things worse and can help the model capture more complex relationships.

Ratio of Height to Weight: The 'Height_to_Weight' ratio is calculated, which could potentially capture information about an individual's body composition and how it relates to cardiovascular health.

Healthy Diet-Lifestyle Interaction: This interaction term, 'HealthyDiet_Lifestyle', represents the product of the healthy diet score and the lifestyle score. This term captures the combined effect of an individual's dietary choices and overall lifestyle habits on their cardiovascular health.

## Converting categorical columns represented as strings into numerical formats to facilitate the modeling process. ##

```python
# Mapping for Diabetes
df_new['Diabetes'] = df_new['Diabetes'].apply(lambda x: 1 if x == 'Yes' else 0)
# Create a new binary column 'Sex_Binary' based on the 'Sex' column
df_new['Sex_Binary'] = df_new['Sex'].apply(lambda x: 1 if x == 'Male' else 0)

# Drop the original 'Sex' column
df_new.drop('Sex', axis=1, inplace=True)


# Convert remaining binary categorical variables to binary format
binary_columns = ['Heart_Disease', 'Skin_Cancer', 'Other_Cancer', 'Depression', 'Arthritis', 'Smoking_History', 'Exercise']
df_new[binary_columns] = df_new[binary_columns].applymap(lambda x: 1 if x == 'Yes' else 0)

    
# Ordinal encoding for General_Health, Checkup, Age_Category
general_health_mapping = {
    'Poor': 1,
    'Fair': 2,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5
}
df_new['General_Health'] = df_new['General_Health'].map(general_health_mapping)

checkup_mapping = {
    'Never': 0,
    '5 or more years ago': 1,
    'Within the past 5 years': 2,
    'Within the past 2 years': 3,
    'Within the past year': 4
    
}
df_new['Checkup'] = df_new['Checkup'].map(checkup_mapping)

age_category_mapping = {
    '18-24': 0,
    '25-29': 1,
    '30-34': 2,
    '35-39': 3,
    '40-44': 4,
    '45-49': 5,
    '50-54': 6,
    '55-59': 7,
    '60-64': 8,
    '65-69': 9,
    '70-74': 10,
    '75-79': 11,
    '80+': 12
}
df_new['Age_Category'] = df_new['Age_Category'].map(age_category_mapping)   
 
# Mapping BMI_category
BMI_Category_mapping = {'Underweight':0, 'Normal weight':1, 'Overweight':2, 'Obesity':3}
  
df_new['BMI_Category'] = df_new['BMI_Category'].map(BMI_Category_mapping) 

# Display the first few rows of the cleaned dataset
df_new
```

The purpose of mapping and encoding is to convert the categorical and ordinal variables in your dataset into a format that can be easily understood by machine learning algorithms.

Diabetes and Sex_Binary: For binary variables like 'Diabetes' and 'Sex', We convert the 'Yes' and 'No' values into 1s and 0s, respectively. This makes it easier for the algorithms to process and understand whether a particular condition or attribute is present or not.

General_Health, Checkup, Age_Category: The values have a specific order or hierarchy. By mapping them to numerical values (like 'Poor' mapped to 1 and 'Excellent' mapped to 5 for 'General_Health'), we were preserving the relative order between the different levels.

BMI_Category: It represents different categories of BMI (Body Mass Index). By mapping the categories to numerical values (e.g., 'Underweight' mapped to 0 and 'Obesity' mapped to 3), creating a numeric representation of the BMI categories.

```python
# Building the base logistic regression model
# Instatiating object
model = LogisticRegression(random_state=42)
# Since we're going to be fitting and evaluating a function would be easier to use instead of repeating the same code again and again.
def fitter_and_evaluator(model, X_train_scaled, X_test_scaled, y_train):
    # Fitting model
    model.fit(X_train_scaled, y_train)
    # Making predictions on test and train
    train_pred = model.predict(X_train_scaled)
    test_pred = model.predict(X_test_scaled)
    # Seeing accuracy, precision, recall and f1_score
    # Accuracy scores
    train_accuracy = accuracy_score(y_train, train_pred)
    test_accuracy = accuracy_score(y_test, test_pred)
    # Precision scores
    train_precision = precision_score(y_train, train_pred)
    test_precision = precision_score(y_test, test_pred)
    # Recall scores
    train_recall = recall_score(y_train, train_pred)
    test_recall = recall_score(y_test, test_pred)
    # F1 scores
    train_f1 = f1_score(y_train, train_pred)
    test_f1 = f1_score(y_test, test_pred)
    # Results
    print('Model accuracy of training data is', round(train_accuracy, 4) * 100,"%")
    print('Model accuracy of testing data is', round(test_accuracy, 4) * 100,"%")
    print('Model precision of training data is', round(train_precision, 4) * 100,"%")
    print('Model precision of testing data is', round(test_precision, 4) * 100,"%")
    print('Model recall of training data is', round(train_recall, 4) * 100,"%")
    print('Model recall of testing data is', round(test_recall, 4) * 100,"%")
    print('Model f1 score of training data is', round(train_f1, 4) * 100,"%")
    print('Model f1 score of testing data is', round(test_f1, 4) * 100,"%")

fitter_and_evaluator(model, X_train_scaled, X_test_scaled, y_train)
    
```

```python
def ROC_curve(X_scaled, y, model):
    """
    Plots ROC_curve given values of X and Y and the model in
    
    Parameters:
       X_scaled: An array of scaled values of the independent variables can be for the training or test dataset.
       y: An array of values of the dependent variables can be for the training or test dataset.
       model: Model used for training and predicting data is usually Logistic Regression model.
    
    Returns:
       ROC_curve of the model
    """
    # Calculate ROC curve
    y_scores = model.predict_proba(X_scaled)[:, 1]
    fpr, tpr, thresholds = roc_curve(y, y_scores)
    
    # Calculate the area under the ROC curve (AUC)
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.figure(figsize=(10, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC)')
    plt.legend(loc='lower right')
    plt.show()

# Call the ROC_curve function with the scaled test data and the trained logistic model
ROC_curve(X_test_scaled, y_test, grid_search)
```

ROC curve of Logistic regression model = 91% 

A 91% ROC curve value suggests that the model has a better ability to correctly classify instances while maintaining a low false positive rate.
